```python
import cv2
```

```python
img1 = cv2.imread('img1.jpg')
```

```python
img1.shape
```

```python
import matplotlib.pyplot as plt
```

```python
plt.imshow(img1)
```

```python
img1 = cv2.cvtColor(img1, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img1)
```

```python
img2 = cv2.imread('img2.png')
```

```python
img2.shape
```

```python
plt.imshow(img2)
```

```python
img2 = cv2.cvtColor(img2, cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img2)
```

```python
img1 = cv2.resize(img1, (1200,1200))
img2 = cv2.resize(img2, (1200,1200))
```

```python
plt.imshow(img1)
```

```python
plt.imshow(img2)
```

# watermark

```python
watermark = cv2.addWeighted(src1=img1, alpha=0.9, src2=img2, beta=0.7, gamma=0)
```

```python
plt.imshow(watermark)
```
